```python
##Qn2
###clustering is used to make clusters of various stars, directors and writers in one frame of similar type
 
```

```python
x=data1.iloc[:,[0,5,9,10,]]
```

```python
x
```

```python
x=np.array(x)
x
```

```python

##Importing Kmeans library
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,12):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

```

```python
plt.plot(range(1,12),wcss)
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss(within sum of squares)")
plt.show()
```

```python
Kmeans=KMeans(n_clusters=10,init='k-means++',random_state=0)
y_kmeans=kmeans.fit_predict(x)
y_kmeans
```

```python
x[y_kmeans==0,0]###These are high budget values
```

```python
data1['Target']=y_kmeans
```

```python
data1.head(8)
```
